```python
import pymongo
import warnings
import pandas as pd
import numpy as np
```

```python
housing_data = pd.read_csv("RDC_InventoryCoreMetrics_Zip.csv")
housing_data.head()
```

```python
housing_data.columns
```

```python
housing_df = housing_data[["ZipCode", "ZipName", "Median Listing Price"]]
housing_df.head()
```

```python
housing_df.columns
```

```python
housing_df.is_copy = False
warnings.simplefilter(action='ignore', category=FutureWarning)

housing_df.rename(columns={'ZipCode': 'zipcode', 'ZipName': 'zipname', 'Median Listing Price': 'median_listing_price'}, inplace=True)
housing_df.head()
```

```python
housing_df.is_copy = False
warnings.simplefilter(action='ignore', category=FutureWarning)

housing_df['zipcode'] = housing_df['zipcode'].astype(str).replace(",", "")
housing_df.head()
```

```python
zipcode = housing_df.zipcode.astype(str).replace("nan", "0").astype(float).astype(int).head(10)
zipcode
# housing_df.zipcode.unique()
```

```python
zipcode = zipcode.apply(lambda x: '{0:0>5}'.format(x))
zipcode.head(10)
```

```python
housing_df["zipcode"] = zipcode
split_zipname = housing_df["zipname"].str.split(",", n=1, expand=True).copy()
housing_df["city"] = split_zipname[0]
housing_df["state"] = split_zipname[1]

housing_df.head()
```

```python
results_housing_df = housing_df[["zipcode", "city", "state", "median_listing_price"]]
results_housing_df.head()
```

```python
results_housing_df.to_csv("./output/totalhousing.csv")
```

```python
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
db = client.housing_db
collection = db.items
```

```python
result_data = results_housing_df.to_dict(orient='records')

[collection.update_one({'zipcode': record['zipcode'], "city": record["city"], "state": record["state"], "median_listing_price": record["median_listing_price"]}, {'$set': record}, upsert=True) for record in result_data]
```
