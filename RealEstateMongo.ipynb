```python
directory = './output/'
realestate_list = []
file = [record for record in os.listdir(directory)]
# print(file)

for data in file:
    with open(os.path.join(directory, data)) as fname:
        read_file = fname.read()

    soup = bs(read_file, "lxml")

    results = soup.find_all('div', class_='zsg-photo-card-content')
#     print(results)

    for item in results:
        realestate_dict = {}
        if item.find('span', class_='hide') is not None:
            address_list = item.find('span', class_='hide')

            address = address_list.find('span', attrs={'itemprop': 'streetAddress'}).text
            city = address_list.find('span', attrs={'itemprop': 'addressLocality'}).text
            state = address_list.find('span', attrs={'itemprop': 'addressRegion'}).text
            zipcode = address_list.find('span', attrs={'itemprop': 'postalCode'}).text
            
            listing_price = ""            
            if item.find('span', class_='zsg-photo-card-price') is not None:
                listing_price = item.find('span', class_='zsg-photo-card-price').text

            realestate_dict = [{"address": address},{"city": city},{"state": state},{"zipcode": zipcode},{"listing_price": listing_price}]

            if not "/mo" in listing_price:
    #             print(listing_price)
    #             append to the list of dictionary
                realestate_list.append(realestate_dict)
    
print(realestate_list)
```

```python
address = pd.DataFrame([i[0] for i in realestate_list])
city = pd.DataFrame([i[1] for i in realestate_list])
state = pd.DataFrame([i[2] for i in realestate_list])
zipcode = pd.DataFrame([i[3] for i in realestate_list])
listing_price = pd.DataFrame([i[4] for i in realestate_list])

realestate_df = pd.concat([address, city, state, zipcode, listing_price], axis=1, sort=True)
realestate_df.head()
```

```python
realestate_df['listing_price'] = realestate_df['listing_price'].str.replace(',', '')
realestate_df['listing_price'] = realestate_df['listing_price'].str.replace('$', '')
realestate_df['listing_price'] = realestate_df['listing_price'].str.replace('K', '')
realestate_df['listing_price'] = realestate_df['listing_price'].str.replace('+', '')
realestate_df.head()
```

```python
realestate_df['listing_price'].replace('', np.nan, inplace=True)
realestate_df.head()
# realestate_df.listing_price.unique()
```

```python
realestate_df.dropna(subset=['listing_price'], inplace=True)
realestate_df.head()
# realestate_df.listing_price.unique()
```

```python
realestate_df['listing_price'] = realestate_df['listing_price'].astype(int)
realestate_df.head()
```

```python
realestate_df.groupby("zipcode", as_index=False).median().head(10)
```

```python
realestate_df.rename(columns={'listing_price': 'median_listing_price'}, inplace=True)
realestate_df.head()
```

```python
realestate_df = realestate_df[["zipcode", "city", "state", "median_listing_price"]]
realestate_df.head()
```

```python
realestate_df.to_csv("./output/zillowdata.csv")
```
