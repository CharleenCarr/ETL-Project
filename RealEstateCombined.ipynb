```python
import pymongo
import warnings
import pandas as pd
from pandas import DataFrame
```

```python
from pymongo import MongoClient
client = MongoClient()
```

```python
client = MongoClient('localhost', 27017)
```

```python
db = client.housing_db
```

```python
collection = db.items
```

```python
housing_locations = collection.find_one()
housing_locations
```

```python
# Expand the cursor and construct the DataFrame
df_1 =  pd.DataFrame(list(collection.find()))
df_1.head()
```

```python
db = client.zillow_db
```

```python
collection = db.items
```

```python
zillow_locations = collection.find_one()
zillow_locations
```

```python
# Expand the cursor and construct the DataFrame
df_2 =  pd.DataFrame(list(collection.find()))
df_2.head()
```

```python
realestate_combined_data = pd.merge(df_1, df_2, on=["zipcode"])
realestate_combined_data.head()
```

```python
# realestate_combined_data.columns
realestate_combined_data.count()
```

```python
realestate_combined_data.rename(columns={'zipcode': 'Combined Zipcode Data', 'city_x': 'US Housing - City', 'state_x': 'US Housing - State', 'median_listing_price_x': 'US Housing - Median Listing Price', 'city_y': 'Zillow - City', 'state_y': 'Zillow - State', 'median_listing_price_y': 'Zillow - Median Listing Price'}, inplace=True)
realestate_combined_data.head()
```

```python
realestate_df = realestate_combined_data[["Combined Zipcode Data", "US Housing - City", "Zillow - City", "US Housing - State", "Zillow - State", "US Housing - Median Listing Price", "Zillow - Median Listing Price"]]
realestate_df.head()
```

```python
realestate_df['Difference in Median Price'] = realestate_df['US Housing - Median Listing Price'].sub(realestate_df['Zillow - Median Listing Price'], axis = 0) 
realestate_df.head()
```

```python
realestate_df.is_copy = False
warnings.simplefilter(action='ignore', category=FutureWarning)

realestate_df.head()
```

```python
realestate_df.to_csv("./output/combinedhousingdata.csv")
```
